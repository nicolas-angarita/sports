```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import json

import time
import random
```

```python
# create the list of seasons to download 
seasons = [str(season) for season in range (2015,2023)]
print(f'number of seasons = {len(seasons)}')

# create the list of team abbreviations
team_abbrs = ['crd','atl','rav','buf','car','chi','cin','cle','dal','den','det','gnb','htx','clt','jax','kan',
              'sdg','ram','rai','mia','min','nwe','nor','nyg','nyj','phi','pit','sea','sfo','tam','oti','was']

print(f'number of teams = {len(team_abbrs)}')
```

```python
seasons = [str(season) for season in range(2000,2001)]
print(f'number of seasons = {len(seasons)}')

weeks = [str(week) for week in range (1,18)]
print(f'number of weeks = {len(weeks)}')
```

```python
#create an empty df to append
nfl_scores = pd.DataFrame()

#iterate through the seasons
for season in seasons:
    #iterate theough the weeks
    for week in weeks:
        #URL set up
        url = f'https://www.pro-football-reference.com/years/{season}/week_{week}.htm'
        print(url)
        
        try:
            #Get the score of the games
            scores_df = pd.read_html(url, header = 1, attrs={'class':'game_summaries'})[0]
            
            nfl_scores = pd.concat([nfl_scores, scores_df], ignore_index=True)
        
            time.sleep(random.randint(7,8))
            
        except ValueError as e:
            print("No tables found on the page:", e)
            
            
print(nfl_scores)            
            
```

```python
#create an empty df to append
nfl_df = pd.DataFrame()

#iterate through the seasons
for season in seasons:
    #iterate through the teams
    for team in team_abbrs:
        #Step 1. Set the URL
        url = 'https://www.pro-football-reference.com/teams/' + team + '/' + season + '/gamelog/'
        print(url)
        
        try:
            #Step 2. Get offensive stats (from game logs table)
            off_df = pd.read_html(url, header=1, attrs={'id':'gamelog' + season})[0]

            #Step 3. Get the defensive stats (from opponents Game logs table)
            def_df = pd.read_html(url, header=1, attrs ={'id': 'gamelog_opp' + season})[0]

            #Step 4. Concat the two dataframes (Offensive and defensive) (along columns, axis = 1)
            team_df = pd.concat([off_df, def_df], axis =1)

            #Insert the season to the df
            team_df.insert(loc=0, column = 'season', value= season)

            #Insert the team to the df
            team_df.insert(loc=1, column = 'team', value = team.upper())

            #Step 5. Concat the team gamelog to the aggregate df (along rows, axis = 0)
            nfl_df = pd.concat([nfl_df,team_df], ignore_index=True)

            #pause the program to abide by the website rules 
            time.sleep(random.randint(7,8))        
        
        except ValueError as e:
            print("No tables found on the page:", e)
            
            
print(nfl_df)        
        
```

```python
# Create an empty DataFrame to append
nfl_scores = pd.DataFrame(columns=['Team A', 'Score A', 'Team B', 'Score B'])

# Define the range of seasons and weeks
seasons = range(2000, 2001)  # Update this as needed
weeks = range(1, 18)  # Update this if the number of weeks changes

# Function to fetch and parse data from a URL
def fetch_scores(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        table = soup.find('table', {'class': 'teams'})
        if table:
            df_list = pd.read_html(str(table), header=1)
            if df_list:
                df = df_list[0]
                return process_table(df)
    return pd.DataFrame(columns=['Team A', 'Score A', 'Team B', 'Score B'])

# Function to process and format the DataFrame
def process_table(df):
    # Assuming the table is in a format where each row represents a game
    rows = []
    for i in range(0, len(df), 2):  # Iterate through the rows in pairs
        if i + 1 < len(df):  # Check to avoid index errors
            row = df.iloc[i:i+2]
            # Extracting relevant columns and formatting
            team_a = row.iloc[0, 0]
            score_a = row.iloc[0, 1]
            team_b = row.iloc[1, 0]
            score_b = row.iloc[1, 1]
            rows.append([team_a, score_a, team_b, score_b])
    return pd.DataFrame(rows, columns=['Team A', 'Score A', 'Team B', 'Score B'])

# Iterate through the seasons
for season in seasons:
    # Iterate through the weeks
    for week in weeks:
        # URL setup
        url = f'https://www.pro-football-reference.com/years/{season}/week_{week}.htm'
        print(url)
        
        try:
            # Get the score of the games
            scores_df = fetch_scores(url)
            
            if not scores_df.empty:
                nfl_scores = pd.concat([nfl_scores, scores_df], ignore_index=True)
            
            time.sleep(random.randint(7, 8))
            
        except Exception as e:
            print("An error occurred:", e)

# Save the DataFrame to a CSV file
nfl_scores.to_csv('scores_nfl.csv', index=False)

# Optional: Verify by reading the CSV file
df_check = pd.read_csv('scores_nfl.csv')
print(df_check.head())
```
